```python
df1 = pd.read_csv(
    "../data/train/my-alt-190530-data",
    sep="\t",
    header=None,
    names=["id", "sentence"]
)
df1.head()
```

```python
with open("../data/train/mypos-ver.3.0.shuf.notag.nopunc.txt", encoding="utf-8") as f:
    lines = f.read().splitlines()
df2 = pd.DataFrame({"text": lines})
df2.head()
```

```python
df3 = load_dataset("URajinda/myanmar_spoken_corpus_v4_cleaned")
```

```python
list(df3["train"].take(5))
```

```python
df3
```

### Remove Uncessary Data

- The first dataset contains an unnecessary column: `id`.
- The third dataset is in dictionary format, so it has to be converted into a dataframe.

```python
df1_new = df1.drop(columns=["id"])
df1_new = df1_new.rename(columns={"sentence": "text"})  # optional
df1_new.head()
```

```python
df3_new = df3["train"].to_pandas()
df3_new.head()
```

## Merge All Datasets

```python
print(df1_new.shape)
print(df2.shape)
print(df3_new.shape)
```

```python
df = pd.concat([df1_new, df2, df3_new], ignore_index=True)
df.head()
```

```python
print(df.shape)
```

```python
df["text"].to_csv("../data/train/combined_raw.txt", index=False, header=False, lineterminator="\n")
```
